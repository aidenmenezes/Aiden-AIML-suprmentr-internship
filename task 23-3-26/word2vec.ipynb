```python
from gensim.models import Word2Vec
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

sentences = [
    "I love watching movies on weekends",
    "The movie was very exciting and fun",
    "That film had amazing visuals",
    "The actor gave a brilliant performance",
    "I enjoy action and thriller movies",
    "The story of the movie was boring",
    "Comedy movies make me laugh",
    "The director did a great job",
    "I like watching movies with friends",
    "The movie had great music and songs",
    "Horror movies are scary but interesting",
    "The film had a weak storyline",
    "I prefer romantic movies sometimes",
    "The movie was too long and slow",
    "The acting in the film was fantastic",
    "The cinema experience was wonderful",
    "This film has a strong storyline",
    "The director created a masterpiece",
    "Actors performed very well in the movie",
    "The movie plot was interesting",
    "I enjoy watching films in theater",
    "The film was full of action scenes",
    "The movie had excellent cinematography",
    "The story and acting were great",
    "The film was entertaining and fun",
    "Movies can be inspiring and emotional",
    "The movie had a powerful message",
    "Film industry produces amazing content",
    "Watching movies is my favorite hobby",
    "The movie had great direction and acting"
]

stop_words = set(stopwords.words('english'))

processed_sentences = []
for sentence in sentences:
    sentence = sentence.lower()
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(sentence)
    tokens = [word for word in tokens if word not in stop_words]
    processed_sentences.append(tokens)

model = Word2Vec(
    sentences=processed_sentences,
    vector_size=100,  
    window=5,        
    min_count=1,
    sg=1               
)

print("Vector for 'movie':\n", model.wv['movie'])

print("\nMost similar words to 'movie':")
for word, score in model.wv.most_similar('movie', topn=5):
    print(word, ":", round(score, 3))
```
