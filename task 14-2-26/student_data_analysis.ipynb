```python
import pandas as pd
df = pd.read_csv("students.csv")

print("Original Data:\n", df)
df["maths"].fillna(df["maths"].mean(), inplace=True)
df["science"].fillna(df["science"].mean(), inplace=True)
df["english"].fillna(df["english"].mean(), inplace=True)

df["total"] = df["maths"] + df["science"] + df["english"]
df["average"] = df["total"] / 3
top3 = df.sort_values(by="total", ascending=False).head(3)
print("\nTop 3 Students:\n", top3)

dept_avg = df.groupby("department")[["maths", "science", "english"]].mean()
print("\nDepartment-wise Average Marks:\n", dept_avg)

high_scorers = df[
    (df["maths"] > 75) &
    (df["science"] > 75) &
    (df["english"] > 75)
]

print("\nStudents scoring above 75 in all subjects:\n", high_scorers)
print("\nFinal Dataset:\n", df)
```
