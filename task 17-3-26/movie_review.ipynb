```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression

data = {
    "review": [
        "The movie was amazing and full of suspense",
        "I loved the acting and storyline",
        "Fantastic film, highly recommended",
        "It was a great experience watching this movie",
        "Absolutely brilliant performance",
        
        "The movie was boring and too long",
        "I did not like the storyline",
        "Worst movie I have ever seen",
        "The acting was terrible",
        "Very disappointing film",
        
        "The movie was okay, not great",
        "It was an average experience",
        "Nothing special but not bad",
        "The film was fine",
        "It was just normal",
        
        "Amazing visuals and great music",
        "Poor direction and weak script",
        "Decent acting but slow story",
        "Loved the cinematography",
        "Not interesting at all",
        "Good but could be better"
    ],
    
    "label": [
        "Positive","Positive","Positive","Positive","Positive",
        "Negative","Negative","Negative","Negative","Negative",
        "Neutral","Neutral","Neutral","Neutral","Neutral",
        "Positive","Negative","Neutral","Positive","Negative","Neutral"
    ]
}

df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(
    df["review"], df["label"], test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

predictions = model.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n", classification_report(y_test, predictions, zero_division=0))

new_review = ["The movie was fantastic and exciting"]
new_review_tfidf = vectorizer.transform(new_review)

prediction = model.predict(new_review_tfidf)
print("\nPrediction for new review:", prediction[0])
```
