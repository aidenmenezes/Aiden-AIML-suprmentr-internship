```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"],
    "Revenue": [20000, 25000, 30000, 28000, 35000, 40000, 42000, 45000],
    "Marketing_Spent": [5000, 7000, 8000, 7500, 9000, 10000, 11000, 12000],
    "Profit": [8000, 10000, 12000, 11000, 15000, 18000, 20000, 22000]
}

df = pd.DataFrame(data)
print("Dataset:\n", df)

plt.figure()
plt.plot(df["Month"], df["Revenue"], marker='o')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

plt.figure()
plt.scatter(df["Marketing_Spent"], df["Profit"])
plt.title("Marketing Spend vs Profit")
plt.xlabel("Marketing Spent")
plt.ylabel("Profit")
plt.show()

correlation = df[["Revenue", "Marketing_Spent", "Profit"]].corr()
print("\nCorrelation Matrix:\n", correlation)

plt.figure()
plt.imshow(correlation)
plt.xticks(range(len(correlation.columns)), correlation.columns)
plt.yticks(range(len(correlation.columns)), correlation.columns)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
```

Insights
Revenue is increasing month by month, which shows the business is growing steadily.
There is a clear relationship between marketing and profit, when marketing spending increases, profit also increases.
Profit follows a similar trend as revenue, meaning higher revenue leads to higher profit.
The data suggests that marketing is helping improve both revenue and profit.
Overall, the business performance is improving over time with consistent growth.
