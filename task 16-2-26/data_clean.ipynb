```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data = {
    "age": [25, 30, np.nan, 22, 30, 25],
    "salary": [50000, np.nan, 60000, 45000, 50000, 50000],
    "city": ["hyderabad", "HYDERABAD", "Hyderabad", "mumbai", "Mumbai", "hyderabad"],
    "experience": [2, 5, 3, 1, 5, 2]
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df)

df = df.drop_duplicates()

df["age"].fillna(df["age"].mean(), inplace=True)
df["salary"].fillna(df["salary"].mean(), inplace=True)

df["city"] = df["city"].str.lower().str.capitalize()

scaler = MinMaxScaler()
df[["age", "salary"]] = scaler.fit_transform(df[["age", "salary"]])

print("\nFinal Cleaned Data:\n")
print(df)
```
